```python
# Databricks notebook source
# 01_init_lakehouse_structure

# =====================================
# 1 Create catalog
# =====================================

spark.sql("CREATE CATALOG IF NOT EXISTS workspace")

# Set the current catalog to 'workspace'
spark.sql("USE CATALOG workspace")

# =====================================
# 2 Create schemas (lakehouse layers)
# =====================================
spark.sql("CREATE SCHEMA IF NOT EXISTS 00_landing")
spark.sql("CREATE SCHEMA IF NOT EXISTS 01_bronze")
spark.sql("CREATE SCHEMA IF NOT EXISTS 02_silver")
spark.sql("CREATE SCHEMA IF NOT EXISTS 03_gold")

spark.sql("SHOW CATALOGS")
# =====================================
# 3 Create landing volume
# =====================================
spark.sql("CREATE VOLUME IF NOT EXISTS 00_landing.landing")
# =====================================
# 4 Bronze tables
# =====================================

spark.sql("""
CREATE TABLE IF NOT EXISTS 01_bronze.bronze_web_events (
    event_id STRING,
    event_time TIMESTAMP,
    session_id STRING,
    user_id STRING,
    event_type STRING,
    product_id STRING,
    order_id STRING,
    quantity INT,
    price_at_event DOUBLE,
    page_url STRING,
    referrer STRING,
    utm_source STRING,
    utm_medium STRING,
    utm_campaign STRING,
    campaign_id STRING,
    device STRING,
    browser STRING,
    store_channel STRING,
    ingestion_time TIMESTAMP,
    source_file STRING
)
USING DELTA
""")

spark.sql("""
CREATE TABLE IF NOT EXISTS 01_bronze.bronze_ads_campaigns (
    date DATE,
    campaign_id STRING,
    campaign_name STRING,
    channel STRING,
    source STRING,
    medium STRING,
    impressions INT,
    clicks INT,
    conversions INT,
    cost DOUBLE,
    currency STRING,
    ingestion_time TIMESTAMP,
    source_api STRING
)
USING DELTA
""")

# print("Lakehouse structure created successfully.")
```
